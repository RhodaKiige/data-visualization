```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars= pd.read_excel("Car data saved.xlsx")
cars.head()
```

```python
# to create scatter plot using seaborn
variables = ['Min Price (Lakh)','Max Price (Lakh)','Range (kmpl)', 'CC','Seats','Variants','Type','Ex-Showroom Price','RTO','Insurance','Other','Onroad Price']
```

```python
plt.figure(figsize =(15,12))
for i, var in enumerate(variables,start=1):# the work enumerate provides both index and value(var) starting at index 1 
    plt.subplot(4,3,i)# to specify the shape of the scatterplots that is three columns and 4 rows.
    sns.scatterplot(x=cars[var], y=cars['Max Price (Lakh)'], alpha=0.7)
    plt.title(f'Scatter plot of {var} vs Onroad Price')
    plt.xlabel(var)
    plt.ylabel("Target")
plt.tight_layout()
plt.show()
```

```python
# histogram for all independent variables
plt.figure(figsize=(15, 10))
for i, var in enumerate(variables, start=1):
    plt.subplot(3,4,i)
    sns.histplot(cars[var], bins=20, kde=True, color="Pink", edgecolor= "Blue")
    plt.title (f"Histogram of {var}")
    plt.xlabel(var)
    plt.ylabel("frequency")
plt.tight_layout()
plt.show()
```

```python
selected_columns=['Min Price (Lakh)', 'Max Price (Lakh)', 'Range (kmpl)', 'CC',
       'Seats', 'Variants', 'Ex-Showroom Price', 'RTO', 'Insurance',
       'Other']
df_selected=cars[selected_columns]
df_selected.head()
```

```python
# to calculate correlation between columns
corr=df_selected.corr()
corr
```

```python
sns.heatmap(corr, cmap='Greens')
plt.title("Heatmap of correlation matrix")
plt.show()
```

```python
sns.heatmap(corr, annot=True, cmap= "BuPu")
plt.title("Heatmap of correlation matrix")
plt.show()
```
