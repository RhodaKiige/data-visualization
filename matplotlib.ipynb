```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('DATAA.XLSX')
df.head()
```

```python
df["Date"] = pd.to_datetime(df["Date"])# to convert date to datetime format
# plotting the data

plt.plot(df["Date"], df["Sales"], color = "Purple")
plt.title ("Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.bar(df["Product"], df["Sales"], color="Purple")
plt.title ("Sales per Product")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.xticks(rotation=0)
plt.show()
```

```python
cars=pd.read_excel("Car data saved.xlsx")
```

```python
cars.head()
```

```python
import numpy as np
plt.figure (figsize=(5,6))
plt.scatter (cars["Max Price (Lakh)"], cars["Insurance"], color="Pink")
plt.title("maxprice(lakh) vs insurance")
plt.xlabel ("max price")
plt.ylabel ("Insurance")
plt.grid (True)
plt.show()
```

```python
import numpy as np
plt.figure (figsize=(5,6))
plt.scatter (cars["Max Price (Lakh)"], cars["Insurance"], color="Pink", label = "data points")
m,b= np.polyfit(cars["Max Price (Lakh)"], cars["Insurance"],1)
plt.plot(cars["Max Price (Lakh)"], m *cars ["Max Price (Lakh)"]+b, color="blue", label ="Trend Line")


plt.title("maxprice(lakh) vs insurance")
plt.xlabel ("max price")
plt.ylabel ("Insurance")
plt.grid (True)
plt.show()
```

```python
# the first thing is to aggregate the data

sales_data= df.groupby ("Category") ["Sales"].sum()
sales_data
```

```python
# to create a pie chart
plt.figure(figsize=(8,8))
plt.pie(sales_data, labels=sales_data.index, autopct='%1.1f%%') #this is to round off

plt.title('Proportion of sales by category')
plt.show()
```
